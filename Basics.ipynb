```python
col_name = ["symboling", "normalized_losses", "make" "fuel_type", "aspiration", "num_doors", "body_style", "drive_wheels", "drive_wheels", "engine_location", "wheel_base", "length", "width", "height", "curb_weight", "engine_type", "num_cylinders", "engine_size", "fuel_system", "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]
data = pd.read_csv("imports-85.data", names = col_name)
data
```

```python
continuous_values_cols = ["normalized_losses", "wheel_base", "length", "width" ,"height", "curb_weight", "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]
numeric_cars = data[continuous_values_cols]
numeric_cars.head(5)
```

```python
import numpy as np
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars
```

```python
numeric_cars = numeric_cars.astype("float")
numerice_cars.isnull().sum()
```

```python
numeric_cars = numeric_cars.dropna(subset=['price'])
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

```python
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
numeric_cars.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
def knn_train_test(feature, target, df, k):
    train = df.iloc[0:150]
    test = df.iloc[150:]
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(train[[feature]], train[target])
    prediction = knn.predict(test[[feature]])
    rmse = mean_squared_error(test[target], prediction) ** (1/2)
    return rmse
cols = numeric_cars.columns.tolist()
cols.remove("price")
test = {}

for i in cols:
    test[i] = knn_train_test(i, 'price', numeric_cars,5)
import operator
test_sorted = sorted(test.items(), key = operator.itemgetter(1))
test_sorted
```

```python

error = {}
k_num = [1,3,5,7,9]
for feature in cols:
    for k in k_num:
        error[feature + str(k)] = knn_train_test(feature, "price", numeric_cars, k)
error_sorted = sorted(error.items(), key=operator.itemgetter(1))
error_sorted
```

```python
def multiple_col(feature, target, df, k):
    train = df.iloc[0:150]
    test = df.iloc[150:]
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(train[feature], train[target])
    prediction = knn.predict(test[feature])
    rmse = mean_squared_error(test[target], prediction) ** (1/2)
    return rmse
two_feature = multiple_col(["curb_weight", "highway_mpg"], "price", numeric_cars, 5)
three_feature = multiple_col(["curb_weight", "highway_mpg", "city_mpg"], "price", numeric_cars, 5)
four_feature = multiple_col(["curb_weight", "highway_mpg", "city_mpg", "width"], "price", numeric_cars, 5)
five_feature = multiple_col(["curb_weight", "highway_mpg", "city_mpg", "width", "horsepower"], "price", numeric_cars, 5)
print(two_feature)
print(three_feature)
print(four_feature)
print(five_feature)
```

```python
outcome = {}
def multiple_col(feature, target):
    np.random.seed(1)
    shuffled_index = np.random.permutation(numeric_cars.index)
    rand = numeric_cars.reindex(shuffled_index)

    train = rand.iloc[0:150]
    test =rand.iloc[150:]
    rmse_val = {}
    for k in range(1,26):
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit(train[feature], train[target])
        prediction = knn.predict(test[feature])
        rmse_val[k] = mean_squared_error(test[target], prediction) ** (1/2)
    return rmse_val
outcome["three feature"] = multiple_col( ["curb_weight", "highway_mpg", "city_mpg"], "price")
outcome["four feature"] = multiple_col(["curb_weight", "highway_mpg", "city_mpg", "width"], "price")
outcome["five feature"] = multiple_col(["curb_weight", "highway_mpg", "city_mpg", "width", "horsepower"], "price")

print(outcome)
import matplotlib.pyplot as plt
%matplotlib inline

for i, k in outcome.items():
    x = list(k.keys())
    y = list(k.values())
    plt.scatter(x,y)
    plt.show()
```
